```python
import torch
from diffusers import StableDiffusionPipeline



model_id = "CompVis/stable-diffusion-v1-4"
device = "cpu"


pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float32)
pipe = pipe.to(device)
```

```python
from IPython.display import display
from typing import Any


def show_generated_image(pipe, prompt: str) -> Any:
    print("="*60)
    print(prompt)
    image = pipe(prompt).images[0]
    display(image)


prompts = [
    "saluting general waste",
    "star wars set in medieval ages",
    "dinosaurs working in a corporation",
    "john carpenters thing set in Poland",
]

for prompt in prompts:
    show_generated_image(pipe, prompt)
```
